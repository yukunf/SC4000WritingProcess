```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv("../data/train_logs_extracted_text.csv")

save_dir = "../data_visualization/graph"
os.makedirs(save_dir, exist_ok=True)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["len_text"], bins=30, kde=True)
plt.title("Distribution of Text Length")
plt.xlabel("Text Length")
plt.ylabel("Count")
plt.savefig(os.path.join(save_dir, "len_text_dist.png"), dpi=300, bbox_inches="tight")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.histplot(df["sentence_count"], bins=20, ax=axes[0], kde=True)
axes[0].set_title("Sentence Count Distribution")

sns.histplot(df["paragraph_count"], bins=20, ax=axes[1], kde=True)
axes[1].set_title("Paragraph Count Distribution")

for ax in axes:
    ax.set_xlabel("")
    ax.set_ylabel("Count")

plt.tight_layout()
plt.savefig(os.path.join(save_dir, "sentence_para_dist.png"), dpi=300, bbox_inches="tight")
plt.show()
```
