```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/train_scores.csv")
df.head()
```

```python
df['score'].describe()
```

```python
import numpy as np
bins = np.arange(0.5, 6.5, 0.5)

plt.figure(figsize=(8,5))
sns.histplot(df['score'], bins=bins, kde=True)
plt.title("Distribution of Writing Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.savefig("../data_visualization/graph/train_score_distribution.png")
plt.show()
```
